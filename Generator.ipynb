```python
Wallet = []
top_up_credit = []
top_up_debit = []
top_up_bank = []
start_date = datetime.date(2019, 1, 1)
end_date = datetime.date(2022, 3, 17)
for x in range(0,2000):
    time_between_dates = end_date - start_date
    days_between_dates = time_between_dates.days
    random_number_of_days = random.randrange(days_between_dates)
    random_date = start_date + datetime.timedelta(days=random_number_of_days)

    credit_card = str(random.choice(credit_card_num_list))
    debit_card = str(random.choice(debitdf_num_list))
    Bank_ID = str(random.choice(bookbank_num_list))

    top_up_credit.append(["CT" + str(x+1).zfill(8), str(round(random.uniform(100,1000000),2)), credit_card,
                        "W"+str(x+1).zfill(9),str(random_date)])
    
    top_up_debit.append(["DT" + str(x+1).zfill(8), str(round(random.uniform(100,1000000),2)), debit_card,
                        "W"+str(x+1).zfill(9),str(random_date)])
    
    top_up_bank.append(["BT" + str(x+1).zfill(8), str(round(random.uniform(100,1000000),2)), Bank_ID,
                        "W"+str(x+1).zfill(9),str(random_date)])

    Wallet.append(["W"+str(x+1).zfill(9), str(round(random.uniform(100,1000000),2)),owner_id_list[x],credit_card,
                    debit_card,Bank_ID])
```

```python
walletdf = pd.DataFrame(Wallet,columns=['Wallet_id','Balanced','Owner_ID','Credit_Card_ID','Debit_Card_ID','Bank_Number'])
credit_topdf = pd.DataFrame(top_up_credit, columns=['Top_up_Credit_Card_ID','Balanced','Credit_Card_Number','Wallet_ID','Top_Up_Date'])
debit_topdf = pd.DataFrame(top_up_debit, columns=['Top_up_Dedit_Card_ID','Balanced','Debit_Card_Number','Wallet_ID','Top_Up_Date'])
bank_topdf = pd.DataFrame(top_up_bank, columns=['Top_up_BankNum_ID','Balanced','Bank_Num','Wallet_ID','Top_Up_Date'])
```

```python
wallet_id_list  = walletdf['Wallet_id'].to_list()
```

```python
transfer = []
start_date = datetime.date(2019, 1, 1)
end_date = datetime.date(2022, 3, 17)

for x in range(0, 1000):
    wallet_list_in_loop = wallet_id_list
    wallet_sender = random.choice(wallet_list_in_loop)
    wallet_list_in_loop.remove(wallet_sender)
    wallet_recv = random.choice(wallet_list_in_loop)
    time_between_dates = end_date - start_date
    days_between_dates = time_between_dates.days
    random_number_of_days = random.randrange(days_between_dates)
    random_date = start_date + datetime.timedelta(days=random_number_of_days)

    transfer.append(["TR" + str(x).zfill(8), str(wallet_sender) ,str(wallet_recv),str(round(random.uniform(100,1000000),2)),str(random_date)])
```

```python
transferdf = pd.DataFrame(transfer,columns=['Transaction_ID','Sender_ID','Receiver_ID','Balacned','Transfer_Date'])
```
